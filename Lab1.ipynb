### Векторы

```r
# присвоить значение вектора чисел от 1 до 10:
vec <- c(1:10)
vec
```

```r
# извлечь элементы с 4-го по 7-ой (получим тоже вектор)
vec_4_7 <- vec[4:7]
vec_4_7
```

```r
# извлечь элементы вектора по условию: нечетные числа
vec2 <- vec[vec %% 2 == 1]
vec2
```

```r
# булевы векторы
vec3 <- c(1, 4, 6, 18, 9)
vec2 == vec3
```

# LAB1: практика

Создайте вектор `n_countries`, состоящий из следующих значений:
58, 27, 23, 20, 9, 6, 5, 5, 4, 4, 3, 3, 3, 3, 3, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2

```r
n_countries <- c(58, 27, 23, 20, 9, 6, rep(5, 2), rep(4, 2), rep(3, 5), rep(2, 20))
n_countries_manual <- c(58, 27, 23, 20, 9, 6, 5, 5, 4, 4, 3, 3, 3, 3, 3, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2)
all(n_countries == n_countries_manual)
```

Используйте функцию `barplot`, чтобы изобразить столбиковую диаграмму со значениями данного вектора.  

```r
# столбиковая диаграмма: показывает значение какого-то признака по нашему наблюдению
barplot(n_countries)
```

```r
names_countries <- c("English", "French", "Arabic", "Spanish", "Portuguese", "German", "Russian", "Swahili", "Italian", "Malay", "Dutch", "Persian", "Sotho", "Tswana", "Tamil", "Albanian", "Chinese", "Romanian", "Somali", "Turkish", "Aymara", "Berber", "Chichewa", "Greek", "Guarani", "Hindustani", "Quechua", "Korean", "Kurdish", "Rwanda-Rundi", "Swati", "Swedish", "Tigrinya", "Venda", "Bengali")
names_countries
```

Давайте добавим в нее в качестве подписей вектор languagelist со следующими номинальными значениями:
"English", "French", "Arabic", "Spanish", "Portuguese", "German", "Russian", "Swahili", "Italian", "Malay", "Dutch", "Persian", "Sotho", "Tswana", "Tamil", "Albanian", "Chinese", "Romanian", "Somali", "Turkish", "Aymara", "Berber", "Chichewa", "Greek", "Guarani", "Hindustani", "Quechua", "Korean", "Kurdish", "Rwanda-Rundi", "Swati", "Swedish", "Tigrinya", "Venda", "Bengali"

```r
barplot(n_countries, names.arg=names_countries, las=2)
```

Подписи к столбцам добавляет аргумент `names.arg=`.  
Поменяйте цвет столбиков с помощью аргумента `col=...`.

```r
barplot(n_countries, col="steelblue", names.arg=names_countries, las=2)
```

Ну а теперь добавим аргументы:
* `horiz=TRUE` -- для поворота диаграммы на 90%  
* `las=1` -- для поворота названий языков
* `xlab="..."` -- для подписи оси X  
* `main="..."` -- для названия всего графика

```r
# добавила las=1 пораньше
barplot(n_countries, col="steelblue", names.arg=names_countries, las=1,
horiz=T, xlab='Number of countries', main='How many countries share a language?')
```

И еще немного красоты. Укажем масштаб подписей:

```r
barplot(rev(n_countries), col="steelblue", names.arg=rev(names_countries), horiz=T, las=1,
cex.main = 2, cex.sub = 0.7, cex.lab = 1.5, cex.names = 1, xlab="Number of countries",
main="How many countries share a language?", sub = "Source: Wikipedia, 2025")
```

Функция rev() - reverse, значения элементов в обратном порядке

С помощью функции `hist()` постройте гистограмму:

```r
hist(n_countries,
     breaks=27)
```

Добавьте к построенной гистограмме цвет столбиков и подписи.  

```r
hist(n_countries,
     breaks=20,
     col='steelblue',
     main='Languages distribution among cointries',
     xlab='Number of countries',
     cex.main=2,
     cex.lab=1.5)
```
