Даны мальтийские существительные (часть датасета из работы Nieder et al. 2021). Постройте скаттерплот / барплот, отражающий частотность этих слов, покажите распределение данных по категориям цветом и формой, подпишите оси и сам график, измените тему на theme_bw()

```r
library(tidyverse)
```

```r
maltese_set <- read.csv('maltese_ft.csv')
str(maltese_set)
```

```r
unique(maltese_set$Origin)
```

```r
# скаттерплот

maltese_set %>%
   ggplot(aes(Word, Occurrences,
 color=pluralType, shape=Verbal)) +
   geom_point() +
   theme_bw(base_size=20) +
   xlab(' ') +
   ylab('Word Frequency') +
   labs(color = "Plural Type",
 shape = "Verbal")+
   scale_x_discrete(labels = NULL) +
   ggtitle('Word Distribution in Maltese')
```

```r
# барплот
maltese_set %>%
   ggplot(aes(Word, Occurrences,
 fill=pluralType)) +
   geom_col() +
   theme_bw(base_size=20) +
   xlab(' ') +
   ylab('Word Frequency') +
   labs(fill = "Plural Type")+
   scale_x_discrete(labels = NULL) +
   ggtitle('Word Distribution in Maltese')
```
